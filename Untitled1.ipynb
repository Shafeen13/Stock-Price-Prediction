```python
# Machine learning
import datetime
from turtle import color
import warnings
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# For data manipulation
import pandas as pd
import numpy as np
```

```python
# To plot
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
# To ignore warnings
warnings.filterwarnings("ignore")
```

```python
# Read the csv file using read_csv
sp=pd.read_csv(r"C:\Users\Shafeen\Desktop\Stock-Price.csv") 
```

```python
# Changes The Date column as index columns
```

```python
# Create predictor variables
sp['Open-Close'] = sp.Open - sp.Close
sp['High-Low'] = sp.High - sp.Low
```

```python
# Store all predictor variables in a variable X
X = sp[['Open-Close', 'High-Low']]
X.head()
```

```python
# Defining target  variables
y = np.where(sp['Close'].shift(-1) > sp['Close'], 1, 0)
```

```python
split_percentage = 0.8
split = int(split_percentage*len(sp))
```

```python
# Train data set
X_train = X[:split]
y_train = y[:split]
```

```python
# Test data set
X_test = X[split:]
y_test = y[split:]
```

```python
# Support vector classifier
print("\nTraining the SVM model...")
cls = SVC().fit(X_train, y_train)
accuracy = cls.score(X_test, y_test)
```

```python
# Calculate Actual returns
sp['Return'] = sp.Close.pct_change()
sp['Actual_Returns'] = sp['Return'].cumsum()
```

```python
# Calculate Predicted returns
sp['Predicted_Signal'] = cls.predict(X)
sp['Single_Pred_Returns'] = sp.Return * sp.Predicted_Signal.shift(1)
sp['Predicted_Returns'] = sp['Single_Pred_Returns'].cumsum()
```

```python
print("\nSVM Model Accuracy: " + str(round((accuracy*100), 2)) + "%")
plt.plot(sp['Actual_Returns'], label='True Returns')
plt.plot(sp['Predicted_Returns'], label='SVM Returns')
plt.title("Prediction by SVM")
plt.xlabel('Date')
plt.ylabel('Returns')
plt.legend()
plt.show()
```
